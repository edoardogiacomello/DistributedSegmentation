```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
```

```python
data =pd.read_csv('results/bin_chk_balanced_vs_unbalanced_c1.csv', index_col=0)
metric_to_show = 'weighted avg_f1-score'
```

Dimensions: method, mu_1_unbalanced, mu_2_unbalanced, mu_balanced, std, unbalanced_agents

```python
for std, filtered_std in data.groupby(['std']):
    for unb_ags, filtered_ags in filtered_std.groupby(['unbalanced_agents']):
        sb.catplot(data=filtered_ags, x='mu_balanced', hue='method', y=metric_to_show, row='mu_1_unbalanced', col='mu_2_unbalanced', kind='bar')
        plt.subplots_adjust(top=0.9)
        plt.gcf().suptitle('Agents: ' + str(unb_ags) + " Std: " + str(std))
        break
```

```python
for std, filtered in data.groupby(['std']):
    sb.catplot(data=filtered, x='mu_balanced', hue='method', y=metric_to_show, row='mu_1_unbalanced', col='mu_2_unbalanced', kind='bar')
```

```python
sb.set_style('whitegrid')
sb.barplot(data=data, hue='method', y=metric_to_show, x='std')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sb.set_style('whitegrid')
sb.barplot(data=data, hue='method', y=metric_to_show, x='unbalanced_agents')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(32, 12))
sb.catplot(data=data, x='unbalance_ratio', hue='method', y=metric_to_show, row='unbalanced_agents', col='mu_balanced', kind='bar')
```
