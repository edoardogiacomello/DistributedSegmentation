```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data =pd.read_csv('results/bin_chk_balanced_vs_unbalanced_c1.csv', index_col=0)
# parameters_col = ['mu_1_unbalanced', 'mu_2_unbalanced', 'mu_balanced', 'std', 'unbalanced_agents', 'method']
# stats = [c for c in data.columns if c not in parameters_col]
# data = data.set_index(parameters_col)['Label 0_f1-score']
data['unbalance_ratio'] = data['mu_1_unbalanced'] / data['mu_2_unbalanced']
```

```python
sb.set_style('whitegrid')
sb.barplot(data=data, hue='method', y='Label 0_f1-score', x='method')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sb.set_style('whitegrid')
sb.barplot(data=data, hue='method', y='Label 1_f1-score', x='std')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sb.set_style('whitegrid')
sb.barplot(data=data, hue='method', y='Label 0_f1-score', x='unbalanced_agents')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(32, 12))
sb.catplot(data=data, x='unbalance_ratio', hue='method', y='Label 0_f1-score', row='unbalanced_agents', col='mu_balanced', kind='violin')
```
