```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
binary_or_multiclass = 'bin'
template = 'blob'
unbalanced_class = 'c2'
data =pd.read_csv('results/{}_{}_balanced_vs_unbalanced_{}.csv'.format(binary_or_multiclass, template, unbalanced_class), index_col=0)
metric_to_show = 'weighted avg_f1-score'
mu1=0.4
mu2=0.6
data = data.loc[data.mu_1_unbalanced == mu1]
data = data.loc[data.mu_2_unbalanced == mu2]
```

```python
# Show difference in negotiations
def generate_meta_method(method):
    meta = 'other'
    if 'Negotiation' in method:
        meta = 'simple negotiation'
    if '(Weighted)' in method:
        meta = 'weighted negotiation'
    if '(Mean-Weighted)' in method:
        meta = 'mean-weighted negotiation'
    return meta

#data = data[data['meta_method'].str.contains('negotiation')]
#data['method'].unique()
data['meta_method'] = data['method'].apply(generate_meta_method)
data = data[data['meta_method']!='other']
```

```python
sb.set_style('darkgrid')
sb.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sb.set_palette(sb.color_palette('Paired',11))
g = sb.relplot(data=data, row='std', col='unbalanced_agents',hue='method', style='meta_method', x = 'mu_balanced', y = metric_to_show, 
               style_order = ['simple negotiation','weighted negotiation','mean-weighted negotiation'], ci=None, kind='line',
               markers = True)   
```

```python
binary_or_multiclass = 'bin'
template = 'chk'
unbalanced_class = 'c2'
data =pd.read_csv('results/{}_{}_balanced_vs_unbalanced_{}.csv'.format(binary_or_multiclass, template, unbalanced_class), index_col=0)
metric_to_show = 'weighted avg_f1-score'
mu1=0.4
mu2=0.6
data = data.loc[data.mu_1_unbalanced == mu1]
data = data.loc[data.mu_2_unbalanced == mu2]
data['meta_method'] = 'mean'
data.loc[data['method']=='Majority Voting','meta_method'] = 'voting'
data.loc[data['method']=='Maximum Proposal','meta_method'] = 'max'
data.loc[data['method'].str.contains("Negotiation"),'meta_method'] = 'negotiation'
data.loc[data['method'].str.contains("Weighted"),'meta_method'] = 'weighted'
sb.set_style('darkgrid')
sb.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sb.set_palette(sb.color_palette('Paired',11))
g = sb.relplot(data=data, row='std', col='unbalanced_agents',hue='method', style='meta_method', x = 'mu_balanced', y = metric_to_show, 
               style_order = ['negotiation','max','weighted','mean','voting'], ci=None, kind='line',
               markers = True)   
```
