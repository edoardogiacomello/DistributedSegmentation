```python
import pandas as pd
import seaborn as sns
import glob
import matplotlib.pyplot as plt
```

```python
RESULTS_FOLDER = 'results/brats2015_validation/'
```

```python
# Loading all MRI CSVs
results = pd.DataFrame()
for p in glob.glob(RESULTS_FOLDER + '/csv/*.csv'):
    temp = pd.read_csv(p, index_col=0)
    results = results.append(temp)
```

```python
sns.set(style="whitegrid")
f, ax = plt.subplots()
# Show each observation with a scatterplot
sns.scatterplot(y="Positive_f1-score", hue="method", x="gt_id", data=results)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
